When we are traveling out of station, or if we think to live in hotel, most people get hesitation of living in hotels because of safety and privacy issues. 

And this place solved all those problems called, "OYO Rooms".
So, we will scrap the web data of OYO Rooms for the city Bhubaneswar and check how many hotels and locations we come up having huge popularity amongst all.

We will scrap the following data:

1.   Hotel name
2.   Location
3.   Original Price (in INR)
4.   Discounted Price (in INR)
5.   Discount (in percentage)
6.   User Rating (out of 5)

```python
# Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')

import csv
from bs4 import BeautifulSoup as bs
import requests
```

```python
# OYO rooms url
url = 'https://www.oyorooms.com/search/?checkin=11%2F12%2F2022&checkout=12%2F12%2F2022&city=Bhubaneswar&country=india&coupon=&filters%5Bcity_id%5D=87&guests=2&location=Bhubaneswar%2C%20Odisha%2C%20India&roomConfig%5B%5D=2&rooms=1&searchType=city&showSearchElements=false&sort=&sortOrder='

headers = { 
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36", 
    "X-Amzn-Trace-Id": "Root=1-63956ce4-077f2bd27750e28d3db65397"
    }
# Send request to the url
page = requests.get(url, headers = headers)

# Access the content in the url
soup = bs(page.content, "html.parser")

# Creating a variable to access only a specific sectional data
lists = soup.find_all('div', class_ = 'oyo-row oyo-row--no-spacing listingHotelDescription') 

# context manager for creating a csv
# Inside the context manager, the all statement are sequential.
with open('hotelData.csv', 'w', newline='', encoding='UTF8') as file:
    # writer() method was imported from csv
    writer = csv.writer(file)

    # This will be the header of the csv file
    header = ['Hotel Name', 'Location', 'Original Price', 'Discounted Price', 'Discount', 'User Rating']

    # writerow() method writes items in sequence list separating them by comma character
    writer.writerow(header)

    # for loop which will loop around the "lists" to search the following sectional data by the help of class_ tags in html doc
    for list in lists:
        hotel_name = list.find('h3', class_ = 'listingHotelDescription__hotelName d-textEllipsis').text
        location = list.find('span', class_ = 'u-line--clamp-2').text
        original_price = list.find('span', class_ = 'listingPrice__slashedPrice d-body-lg').text
        discount_price = list.find('span', class_ = 'listingPrice__finalPrice').text
        discount = list.find('span', class_ = 'listingPrice__percentage').text
        rating = list.find('div', class_ = 'hotelRating').text

        # This will be the data for the rows
        data = [hotel_name, location, original_price[1:], discount_price[1:], discount[0:3], rating[0:3]]
        writer.writerow(data)
```
